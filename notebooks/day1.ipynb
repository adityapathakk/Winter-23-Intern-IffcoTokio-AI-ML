```python
# Importing libraries
import matplotlib.pyplot as plt, numpy as np
```

```python
# Generating a random series
x = np.linspace(1,20,1000)
print(x)
```

```python
# Function for Normnal Distribution
def normalDistribution(x,mean,sd):
    probDensity = (np.pi * sd) * np.exp(-0.5 * ((x - mean)/sd)**2)
    return probDensity
```

```python
# Calculating mean and standard deviation of random series
mean = np.mean(x)
sd = np.std(x)
```

```python
# Applying function to data
pdf = normalDistribution(x,mean,sd)
```

```python
# Plotting results
plt.plot(x,pdf)
```
